```python
# test 1
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# ระบุ URL หรือที่อยู่ของรูปภาพ
image_url2 = "/content/drive/MyDrive/image/E1.png"

# โหลดรูปภาพ
image = cv2.imread(image_url2)
height, width, _ = image.shape
print(f"ความกว้าง: {width} พิกเซล")
print(f"ความสูง: {height} พิกเซล")

# เปลี่ยนขนาด
resized_image = cv2.resize(image,(576,384))

cv2_imshow(resized_image)
resized_image.shape

    # ดึงขนาดของรูปภาพ
height, width, _ = resized_image.shape
print(f"ความกว้าง: {width} พิกเซล")
print(f"ความสูง: {height} พิกเซล")
    # แสดงรูปภาพ
img_size=plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # ไม่แสดงแกน X และ Y
plt.show(img_size)


```
